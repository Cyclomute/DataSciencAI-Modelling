### Age Pyramid (Population Pyramid) of the census19 cleaned data

```python
#Creating the age pyramid (population pyramid); source code: stackoverflow
#calculate the min and max age values
min_age = int(df['Age'].min())
max_age = int(df['Age'].max() + 0.1)

# set the bin size and create the bins
bin_size = 5
bins = range(min_age, max_age + bin_size, bin_size)

# create the age groups
df['age_group'] = pd.cut(df['Age'], bins = bins, labels = [f'{i}-{i+bin_size-1}' for i in bins[:-1]])

# group by age group and gender and calculate the total count
age_groups = df.groupby(['age_group', 'Gender']).size().unstack()

# separating the male and female data-frames
males = age_groups['Male'].sort_index(ascending=True)
females = age_groups['Female'].sort_index()

# plotting the age pyramid (population pyramid)
fig, ax = plt.subplots(figsize=(8, 6))

# plotting male and female bars separately
ax.barh(males.index, -males, height=-0.5, align='center', color='green', alpha=0.7, left=0, label='Male')
ax.barh(females.index, females, height=0.5, align='center', color='red', alpha=0.7, label='Female')

# set the x-axis limits and ticks
max_count = max(abs(males).max(), abs(females).max())
ax.set_xlim([-max_count * 1.1, max_count * 1.1])
ax.set_xticks(range(-int(max_count * 1.1), int(max_count * 1.1) + 1, int(max_count * 0.1)))

# add vertical line at x = 0
ax.axvline(x=0, linestyle='-', color='black', alpha=0.5)

# Labeling the x and y axes
ax.set_xlabel('Number of People')
ax.set_ylabel('Age Group')
ax.set_title('Population Pyramid')

# adding the legend
ax.legend(loc='upper right')

#removing the negative sign on the x-axis for Males
xticks = plt.gca().get_xticks().astype(int)
plt.xticks(xticks, labels = np.abs(xticks))

plt.show()
```

### Calculating Fertility Rate, Birth Rate, and Death Rate

```python
#checking for new born babies in the dataframe (i.e. whose ages are zero & one or < 2)
(df["Age"] < 2).value_counts()
```

```python
#checking the sum of the ages of new born babies in the dataframe 
infant_age = (df["Age"] < 2).sum()
infant_age
```

```python
#checking the total female entries
female_count = (df["Gender"] == "Female").sum()
female_count
```

```python
#calculating the fertility rate as total fertility rate (i.e. the average number of children per 100 women)
total_fertility_rate = round((infant_age/female_count)*100, 2)
print("total_fertility_rate = ", total_fertility_rate)
```

```python
# Calculating the potential women of childbearing age, 15 to 45
child_bearing_age = ((df["Age"] >=15) & (df["Age"] <=45) & (df["Gender"] == "Female")).sum()
child_bearing_age
```

```python
# Calculating birth rate (i.e general fertility rate) as birth per 1000
birth_per_1000 = round((infant_age/child_bearing_age)*1000, 2)
print("birth_per_1000 = ", birth_per_1000)
```

```python
# calculating the crude birth rate as birth per 1000 of the population
crude_birth_per_1000 = round((infant_age/population_size)*1000, 2)
crude_birth_per_1000
```

```python
#Obtaining the life expectancy
life_expectancy = (df["Age"] > 80).sum()
life_expectancy 
```

```python
#calculating death rate of the populattion per 1000
death_per_1000 = round((life_expectancy/population_size)*1000, 2)
print("death_per_1000 = ", death_per_1000)
```

```python
# Calculating the population growth, i.e. the difference between the birth rate and death rate
birth_death_delta = round((birth_per_1000 - death_per_1000), 2)
print(f"Population_growth_delta = {birth_death_delta}")
```

### Occupation Analysis

```python
#Calculating unemployed people and rate
unemployed = (df["Occupation"] == "Unemployed").sum()
print("unemplyed:", unemployed)
population = 7761
unemployment_rate = round((unemployed/population)*100, 2)
print("unemployment_rate:", unemployment_rate)
```

```python
### Calculating retired people and rate assuming >65years as retirement age
retired = (df["Age"] >65).sum()
print("retired: ", retired)
population = 7761
retired_rate = round((retired/population)*100, 2)
print("retired rate: ", retired_rate)
```

```python
# Calculating minors (<18years) who should not be working
minors = (df["Age"] < 18).sum()
print("minors: ", minors)
population = 7761
minors_rate = round((minors/population)*100, 2)
print("minors rate:", minors_rate)
```

```python
#Calculating the working-class
all_students = student_commuters_percentage
working_class = 100 - (unemployment_rate + retired_rate + minors_rate + all_students)
print("working-class: ", working_class)
```
