```python
# Importing the required library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Fetching and reading in the data
df = pd.read_csv("/Users/morrisonosiezagha/Downloads/uber_pickups.csv")

# Exploring and displaying the first five roles of the dataset 
print(df.head())

# Preprocessing the data
df.dropna(inplace=True) # Clean data
df["pickup_time"] = pd.to_datetime(df["pickup_time"])
df["hour"] = df["pickup_time"].dt.hour

# Histogram of the distribution of pickups by hour
sns.histplot(df["hour"], bins=24, kde=False)
plt.title("Distribution of Uber Pickups by Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Pickups")
plt.show()

# Scatter plot of the distribution of the number of rides by hour
plt.scatter(df["hour"], df["number_of_rides"])
plt.title("Number of Uber Rides DrisbRuted by Hour")
plt.ylabel("Number of Uber Rides")
plt.xlabel("Hour of the Day")
plt.show()
```
